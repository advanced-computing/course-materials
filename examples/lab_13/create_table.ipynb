```python
# and we will store the data in bigquery
import pandas_gbq
import yfinance as yf  # for downloading stock data
```

### Getting daily data from last month


```python
dat = yf.Ticker("MSFT")
msft_df = dat.history(period="1mo")
# drop last 2 rows, so we can update later
msft_df = msft_df[:-2]
msft_df
```

## Creating BigQuery Table


```python
from load_data import PROJECT_ID, TABLE

# to BigQuery
pandas_gbq.to_gbq(
    msft_df,
    destination_table=TABLE,
    project_id=PROJECT_ID,
    if_exists="replace",
)
```
